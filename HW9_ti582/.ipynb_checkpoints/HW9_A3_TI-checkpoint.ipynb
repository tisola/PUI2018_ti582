```python
# Download the data
FirearmURL = 'https://serv.cusp.nyu.edu/~fbianco/PUIdata/WorldFirearmsMurdersAndOwnership.csv'
FA = pd.read_csv(FirearmURL)
FA.drop(['ISO code','Source','Rank by rate of ownership','Homicide by firearm rate per 100,000 pop','% of homicides by firearm'],axis=1,inplace=True)
FA = FA.rename(index=str, columns={'Country/Territory':'Country'})

msURL = '''data:application/octet-stream;charset=utf-8,Country%2CGuns%20per%20100%20inhabitants%2CNumber%20of%20mass%20shootings%20
(1983-2013)%0AAustralia%2C15%2C4%0AAustria%2C30.4%2C1%0ABelgium%2C17.2%2C2%0ACanada%2C30.8%2C4%0ADenmark%2C12%2C0%0AFinland%2C45.3%
2C2%0AFrance%2C31.2%2C6%0AGermany%2C30.3%2C7%0AGreece%2C22.5%2C0%0AHolland%2C3.9%2C2%0AHungary%2C5.5%2C0%0AIceland%2C30.3%2C0%0AIreland%
2C8.6%2C0%0AItaly%2C11.9%2C0%0AJapan%2C0.6%2C0%0ALuxembourg%2C15.3%2C0%0ANew%20Zealand%2C22.6%2C3%0ANorway%2C31.3%2C1%0APoland%2C1.3%2C0%
0APortugal%2C8.5%2C1%0ASpain%2C10.4%2C1%0ASweden%2C31.6%2C2%0ASwitzerland%2C45.7%2C2%0AUnited%20Kingdom%2C6.2%2C3%0AUnited%20States%2C88.8%2C78
'''
ms = urllib.urlretrieve(msURL,'ms.csv')
# Move the data to the PUIDATA folder
currentPath = os.getcwd()
newPath = os.getenv('PUIDATA')
shutil.move('%s/ms.csv'%currentPath, '%s/ms.csv'%newPath)
ms = pd.read_csv('%s/ms.csv'%newPath)
ms.drop(0,axis=0,inplace=True)
ms.drop(['Guns per 100 inhabitants'],axis=1,inplace=True)
ms = ms.rename(index=str, columns={'Number of mass shootings ':'Number of mass shootings'})

popURL = 'http://api.worldbank.org/v2/en/indicator/SP.POP.TOTL?downloadformat=csv'
if(os.path.isfile('%s/API_SP.POP.TOTL_DS2_en_csv_v2_10224786.csv'%newPath)):
    print("File in place. Continue")
else:    
    urllib.urlretrieve(popURL, "pop.gz")
    !unzip -d $PUIDATA pop.gz
```

```python
# Convert the population data to a pandas dataframe
# The first few rows of the population CSV file are not applicable
usefulRows = np.arange(5,270,1)

with open('%s/API_SP.POP.TOTL_DS2_en_csv_v2_10224786.csv'%newPath) as popcsv:
    reader = csv.reader(popcsv)
    usefulRows2 = [row for index, row in enumerate(reader) if index in usefulRows]

pop = pd.DataFrame(np.array(usefulRows2).reshape(264,63))
cols_to_drop = [i for i in pop.columns.values if i != 0 and i != 56]
pop.drop(cols_to_drop, axis=1, inplace=True)
pop = pop.rename(index=str, columns={0:'Country', 56:'Population'})
```

```python
# Download GDP data
gdpURL = 'http://api.worldbank.org/v2/en/indicator/NY.GDP.MKTP.CD?downloadformat=csv'

if(os.path.isfile('%s/API_NY.GDP.MKTP.CD_DS2_en_csv_v2_10224782.csv'%newPath)):
    print("File in place. Continue")
else:    
    urllib.urlretrieve(gdpURL, "gdp.gz")
    !unzip -d $PUIDATA gdp.gz
```

```python
# Convert the GDP data to a pandas dataframe
# The first few rows of the GDP CSV file are not applicable
usefulRowsG = np.arange(5,270,1)

with open('%s/API_NY.GDP.MKTP.CD_DS2_en_csv_v2_10224782.csv'%newPath) as gdpcsv:
    reader = csv.reader(gdpcsv)
    usefulRowsG2 = [row for index, row in enumerate(reader) if index in usefulRowsG]

gdp = pd.DataFrame(np.array(usefulRowsG2).reshape(264,63))
cols_to_drop = [i for i in gdp.columns.values if i != 0 and i != 56]
gdp.drop(cols_to_drop, axis=1, inplace=True)
gdp = gdp.rename(index=str, columns={0:'Country', 56:'GDP'})
```

```python
merged1 = pd.merge(FA, ms, left_on='Country', right_on='Country', how='inner')
merged2 = pd.merge(merged1, gdp, left_on='Country', right_on='Country', how='inner')
merged = pd.merge(merged2, pop, left_on='Country', right_on='Country', how='inner')
merged[['GDP']] = merged[['GDP']].astype(float)
merged.loc[merged['Country']=='Finland','Number of mass shootings']=0
merged.loc[merged['Country']=='Poland','Number of mass shootings']=0
merged[['Number of mass shootings']] = merged[['Number of mass shootings']].astype(int)
merged[['Population']] = merged[['Population']].astype(int)
merged.head()
```

# Modeling

## Fit a line to the Number of mass shootings per person as a function of Average total all civilian firearms per person.

### OLS

```python
x = merged['Number of mass shootings']
y = merged['Average total all civilian firearms']/1e7
x2 = sm.add_constant(x)
smfit = smf.OLS(y, x2).fit()
smfit.summary()
```

```python
# Plotting
plotOLS = pl.figure(figsize=(7,7)).add_subplot(111);
plotOLS.errorbar(x,y,yerr=np.sqrt(x),fmt='o',label='Errors');
plotOLS.plot(x,smfit.fittedvalues,label='OLS Fitted Line')
plotOLS.set_title('Number of mass shootings vs number of civilian firearms',fontsize=15)
plotOLS.set_xlabel("Number of civilian firearms (in Tens of Millions)")
plotOLS.set_ylabel("Number of mass shootings")
plotOLS.grid(True)
plotOLS.legend()
```

**Caption**: The number of mass shootings vs the number of civilian firearms with a fitted line using the OLS method. The United States has a larger number of firearms and mass shootings.

### WLS

```python
weights = (1/(np.sqrt(x)))
smfitWLS = smf.WLS(y, x2, weights).fit()
smfitWLS.summary()
```

```python
plotWLS = pl.figure(figsize=(7,7)).add_subplot(111);
plotWLS.errorbar(x,y,yerr=np.sqrt(x),fmt='o',label='Errors');
plotWLS.plot(x,smfitWLS.fittedvalues,label='WLS Fitted Line')
plotWLS.set_title('Number of mass shootings vs number of civilian firearms',fontsize=15)
plotWLS.set_xlabel("Number of civilian firearms (in Tens of Millions)")
plotWLS.set_ylabel("Number of mass shootings")
plotWLS.grid(True)
plotWLS.legend()
```

**Caption**: The number of mass shootings vs the number of civilian firearms with a fitted line using the WLS method. The United States has a larger number of firearms and mass shootings.
