## Task 4: Identify periodicity using Fourier transforms

```python
# Convert data to dataframe
mta = mta.astype('float64')
mtaStations = np.nanmean(mta, axis=1)
dates = pd.date_range(start='05/19/2010', end='02/05/2014', freq='W')
mtaStationsDF_withnan = pd.DataFrame(mtaStations)
```

```python
# Drop NaN values
mtaStationsDF = mtaStationsDF_withnan.dropna(axis=0, how='any')
```

```python
# Plot one station as an example
ind = 0
f = np.abs(np.fft.rfft(mtaStationsDF.iloc[ind]))**2

N = len(mtaStationsDF.iloc[ind])
fig = pl.figure(figsize=(15,5))
ax = fig.add_subplot(111)
ax.plot(np.fft.rfftfreq(N, 1.0)[1:], (f[1:]), 'o', ms=10)
ax.plot(np.fft.rfftfreq(N, 1.0)[1:], (f[1:]), '-')
ax.set_xticklabels(["%s"%(1/f) for f in ax.get_xticks()], fontsize=15);
pl.ylabel("Power", fontsize=15);
pl.xlabel("Period (weeks)", fontsize=15);
pl.title("Power vs Period for Station Number %s using Fourier"%(ind+1), fontsize=20);
```

**Figure 1**: A plot of the fourier transformation of station number 1. The data was created by calculating an average number of swipes for 23 different card types for 600 MTA stations. This resulted in an array of 600 stations each with swipe data for 194 weeks starting from 05/19/2010. I plotted the above just to see what one plot would be as an example. I will now proceed to find the 4 stations with an annual periodicity.

```python
# Calculate the prominent points at which the fourier transformation can show a spike
periods = pd.DataFrame()
N = mtaStationsDF.shape[1]
periods['Critical Points'] = 1 / np.fft.rfftfreq(N, 1.0)
periods.head(10)
```

Above is a DataFrame showing the prominent points at which the period values can take for the fourier transformation. I have calculated the inverse of the frequency to show the period in the above DataFrame. More specifically, the above DataFrame shows the period in units of weeks. Since I am interested in MTA stations that have an annual periodicity, I am searching for stations that have a spike around 52 weeks. The indices in the above DataFrame that are closest to 52 are index 3 and index 4. Thus, I will look for MTA stations that have a spike in the fourier transformation at period index 3 or 4. 

```python
# Create an empty list
annualstations = np.array([])
```

```python
# Populate the empty list with the fourier transformations of the MTA stations that have
# their largest spikes at period index 3 or period index 4
for i in arange(0, 417):
    f = np.abs(np.fft.rfft(mtaStationsDF.iloc[i]))**2
    testcondition = (-f).argsort()[0:2]
    if (testcondition == 3).any() or (testcondition == 4).any():
        #print('%s'%i)
        annualstations = np.append(annualstations, i)
annualstations = annualstations.astype('int64')
```

```python
# Plot the stations from the populated list
N = len(mtaStationsDF.iloc[ind])

fig = pl.figure(figsize=(15,5))
ax = fig.add_subplot(111)

for item in annualstations:
    f = np.abs(np.fft.rfft(mtaStationsDF.iloc[item]))**2
    ax.plot(np.fft.rfftfreq(N, 1.0)[1:], (f[1:]), '-')
ax.set_xticklabels(["%s"%(1/f) for f in ax.get_xticks()], fontsize=20)
pl.ylabel("Power", fontsize=15);
pl.xlabel("Period (weeks)", fontsize=15);
pl.title("Power Spectrum for potential annually periodic stations", fontsize=20);
```

**Figure 2**: A plot of the fourier transformation of the potential annually periodic stations. There is clearly a spike occuring at what appears to be 52 weeks but it is hard to tell visually which the 4 most prominent ones are so I will now proceed to calculate them.

```python
# Create a list of the maximum value of each station's power spectrum and the corresponding station number
maxstations = list([])
for item in annualstations:
    f = np.abs(np.fft.rfft(mtaStationsDF.iloc[item]))**2
    maxstations.append({'Max': f.max(), 'StationNum': item})
```

```python
# Sort the list by descending order to identify the 4 most prominent ones
theStations = pd.DataFrame(maxstations)
theStations.sort_values('Max', ascending=False)
```

As calculated above, the four most prominent annually periodic stations are stations 28, 50, 11, and 175. The annual periodicity, in my opinion, may be due to those stations having regular peaks in their usage, such as around holidays when more tourists come to New York City. That could potentially create regular annual periodicity at those stations.
